```python
display(tabela["product_category"].value_counts())


# Categoria mais vendida de produtos por quantidade: beleza e moda
# Categoria menos vendida de produtos por quantidade: livros

# Categorias mais vendidas por valor: esportes e eletrônicos
# Categoria menos vendida por valor: livros

# Regiões com mais vendas: Sudeste e Nordeste.
# Regiões com menos vendas: Norte e Centro-Oeste.

# INSIGHT 1: promover a venda de livros por meio de um evento ou promovendo um desconto.  

# INSIGHT 2: promover o marketing voltado às regiões com menos clientes com foco em aquisição de clientes.  




# Verificando a quantidade vendida por tipo de produto: 

print("Divisão por categoria:")
tabelaProd = tabela.groupby("product_category")["total_value"].sum().reset_index()
tabelaProd = tabelaProd.sort_values("total_value")


display(tabelaProd)


print("-"*50)
print(" "*50)

# Verificando a quantidade vendida por região: 

print("Divisão por regiões:")
tabelaRegion = tabela.groupby("customer_region")["total_value"].sum().reset_index()
tabelaRegion = tabelaRegion.sort_values("total_value")

display(tabelaRegion)
```

```python
tabelaItem = tabela.groupby(["product_name", "product_category"])["total_value"].sum().reset_index()

tabelaItem = tabelaItem.sort_values("total_value", ascending=False)


# Itens mais vendidos

print("Produtos mais vendidos:")
display(tabelaItem.head(10))

print("-"*50)
print(" "*50)

# Itens menos vendidos

print("Produtos menos vendidos:")
display(tabelaItem.tail(10))

```
