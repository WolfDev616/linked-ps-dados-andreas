```python
display(tabela["product_category"].value_counts())


# Categoria mais vendida de produtos: beleza e moda
# Categoria menos vendida de produtos: livros


# INSIGHT 1: oferecer mais livros sobre beleza e moda. 



# Verificando a quantidade vendida por tipo de produto: 

tabela = tabela.groupby("product_category")["total_value"].sum()

display(tabela)

# Verificando a quantidade vendida por região: 

tabela = tabela.groupby("customer_region")["total_value"].sum()

display(tabela)
```

```python
# Itens mais vendidos

tabela = tabela.groupby("product_name")["total_value"].sum()
# tabela.sort_values("total_value")

display(tabela)

```
