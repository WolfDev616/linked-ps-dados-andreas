Projeto de Análise da planilha de vendas de um e-commerce fictício. 

```python
import pandas as pd
import plotly.express as px

tabela = pd.read_csv("/home/andreas/Programação/Projetos Python/Análise de Dados/Linked/vendas_linked_ps.csv")

display(tabela)

base_color = ["green"]
```

```python
# Verificando se a tabela tem valores nulos

display(tabela.info())
```

```python
tabela.describe()

# Valor médio de vendas: R$ 185,36

# Menor valor de produto: R$ 26,28

# Maior valor de produto: R$ 1635,01
```

```python
display(tabela["product_category"].value_counts())

# INFO

# Categoria mais vendida de produtos por quantidade: beleza e moda
# Categoria menos vendida de produtos por quantidade: livros

# Categorias mais vendidas por valor: esportes e eletrônicos
# Categoria menos vendida por valor: livros

# Regiões com mais vendas: Sudeste e Nordeste.
# Regiões com menos vendas: Norte e Centro-Oeste.

# INSIGHT 1: promover a venda de livros por meio de um evento ou oferecendo um desconto.  

# INSIGHT 2: promover o marketing voltado às regiões com menos clientes com foco em aquisição de clientes.




# Verificando a quantidade vendida por tipo de produto: 

print("Divisão por categoria:")
tabelaProd = tabela.groupby("product_category")["total_value"].sum().reset_index()
tabelaProd = tabelaProd.sort_values("total_value")

display(tabelaProd)


# Gráfico tipo de produto vs valor em receita

graficoCategoria = px.bar(tabelaProd, x='product_category', y='total_value', color="product_category", color_discrete_sequence = [base_color])

graficoCategoria.show()




print("-"*200)
print(" "*50)


# Verificando a quantidade vendida por região: 

print("Divisão por regiões:")
tabelaRegion = tabela.groupby("customer_region")["total_value"].sum().reset_index()
tabelaRegion = tabelaRegion.sort_values("total_value")

display(tabelaRegion)


# Gráfico região vs valor em receita:

graficoCategoria = px.bar(tabelaRegion, x='customer_region', y='total_value', color="customer_region", color_discrete_sequence = [base_color])

graficoCategoria.show()
```

```python
tabelaItem = tabela.groupby(["product_name", "product_category"])["total_value"].sum().reset_index()

tabelaItem = tabelaItem.sort_values("total_value", ascending=False)


# Itens mais vendidos

print("Produtos mais vendidos:")
display(tabelaItem.head(10))




# Gráfico itens mais vendidos vs valor em receita:

graficoCategoriaMais = px.bar(tabelaItem.head(10), x='product_name', y='total_value', color="product_name", color_discrete_sequence = [base_color])

graficoCategoriaMais.show()


print("-"*200)
print(" "*50)



# Itens menos vendidos

print("Produtos menos vendidos:")
display(tabelaItem.tail(10))

# Gráfico itens menos vendidos vs valor em receita:

graficoCategoriaMenos = px.bar(tabelaItem.tail(10), x='product_name', y='total_value', color="product_name", color_discrete_sequence = [base_color])

graficoCategoriaMenos.show()
```

```python
# Tabela dos métodos de pagamento vs valor em receita:

tabelaPayment = tabela.groupby("payment_method")["total_value"].sum().reset_index()

display(tabelaPayment)


# Gráfico dos métodos de pagamento vs valor em receita:

graficoPayment = px.pie(tabelaPayment, values='total_value', names='payment_method')

graficoPayment.show()




# INFO:
# Maior quantidade de receita é recebida por mei ode cartão e pix. 
```

```python
tabelaCancel = tabela.groupby("order_status")["total_value"].sum().reset_index()

display(tabelaCancel)

# Porcentagem de cancelamentos

graficoCancel = px.bar(tabelaCancel, x='total_value', y='order_status', color='order_status', color_discrete_sequence = [base_color])

graficoCancel.show()


# INFO

# Valor total das compras canceladas: R$ 38.367,13. 
# INSIGHT 3: É um valor muito alto. Uma possível abordagem seria enviar um e-mail oferecendo um desconto quando a compra for cancelada. 
```

```python
tabelaCancelReg = tabela.groupby("customer_region")["order_status"].value_counts().reset_index()

# Total por região

total_regiao = tabelaCancelReg.groupby("customer_region")["count"].sum()


# Só os cancelados

cancelados = tabelaCancelReg[tabelaCancelReg["order_status"] == "Cancelado"].set_index("customer_region")["count"]


# Porcentagem de cancelamentos

porcentagem = (cancelados / total_regiao * 100).round(2).reset_index()
porcentagem.columns = ["customer_region", "cancelamento_%"]
porcentagem = porcentagem.sort_values("cancelamento_%", ascending=False)
display(porcentagem)


# Porcentagem de cancelamentos por região

graficoCancelReg = px.bar(porcentagem, x='customer_region', y='cancelamento_%', color_discrete_sequence = [base_color])

graficoCancelReg.show()


#INFO

# INSIGHT 4: A região sul é a terceira com menos clientes e a primeira na porcentagem de cancelamentos. Possivelmente fazer uma campanha de descontos. Verificar causas (como prazos de entrega e frete, por exemplo) e trabalhar em cima da informação obtida. Possivelmente fazer uma campanha de frete grátis, por exemplo, mas é necessário entender o porquê dos cancelamentos. 
```

```python
tabelaCancProd = tabela.groupby(["product_category"])["order_status"].value_counts().reset_index()

# display(tabelaCancProd)

total_categoria = tabelaCancProd.groupby("product_category")["count"].sum()

cancelados = tabelaCancProd[tabelaCancProd["order_status"] == "Cancelado"].set_index("product_category")["count"]

porcentagem = (cancelados / total_categoria * 100).round(2).reset_index()
porcentagem.columns = ["product_category", "cancelamento_%"]
porcentagem = porcentagem.sort_values("cancelamento_%", ascending=False)
display(porcentagem)

# Nenhuma informação importante aqui neste ponto
```
